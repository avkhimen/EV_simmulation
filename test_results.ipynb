```python
def compare_Q(Q1, Q2):
    l_2 = []
    for row in Q1:
        l = []
        for column in row:
            max_action = np.argmax(column[0])
            l.append(max_action)
        l_2.append(np.array(l))
    l_Q1 = np.array(l_2)
    
    l_2 = []
    for row in Q2:
        l = []
        for column in row:
            max_action = np.argmax(column[0])
            l.append(max_action)
        l_2.append(np.array(l))
    l_Q2 = np.array(l_2)
    
    return np.count_nonzero(l_Q1-l_Q2)
```
